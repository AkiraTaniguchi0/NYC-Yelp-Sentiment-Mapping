MOST COMMON CATEGORIES

```python
data_pivot= pd.pivot_table(dat, values = 'address1', index=['cluster'], columns=['categories1'], aggfunc="count")
data_pivot.fillna(0, inplace=True)
data_pivot.reset_index(inplace=True)

data2_pivot= pd.pivot_table(dat, values = 'address1', index=['cluster'], columns=['categories2'], aggfunc="count")
data2_pivot.fillna(0, inplace=True)
data2_pivot.reset_index(inplace=True)

data3_pivot= pd.pivot_table(dat, values = 'address1', index=['cluster'], columns=['categories3'], aggfunc="count")
data3_pivot.fillna(0, inplace=True)
data3_pivot.reset_index(inplace=True)

df = data_pivot.set_index('cluster').add(data2_pivot.set_index('cluster'), fill_value=0).reset_index()
df2 = df.set_index('cluster').add(data3_pivot.set_index('cluster'), fill_value=0).reset_index()
```

```python
df2.head()
```

```python
final["category1"] = np.nan
final["category2"] = np.nan
final["category3"] = np.nan
```

```python
for i in range(0, df2.shape[0]):   
    a = df2.iloc[i,1:]
    b = a.sort_values(ascending=False)[:3]

    final['category1'][i] = b.index[0]
    final['category2'][i] = b.index[1]
    final['category3'][i] = b.index[2]
```

AVG PRICE

```python
def flg(dat):
    if (dat['price'] == '$'):
        return 1
    elif (dat['price'] == '$$'):
        return 2
    elif (dat['price'] == '$$$'):
        return 3
    elif (dat['price'] == '$$$$'):
        return 4
    else:
        return np.nan
```

```python
dat['price_new'] = dat.apply(flg, axis=1)
dat2 = dat[['cluster','price_new']]
dat3 = dat2.groupby(['cluster'], as_index=False).mean()
```

```python
dat3
```

```python
def flg2(dat):
    if (dat['price_new'] < 1.5):
        return '$'
    elif (dat['price_new'] < 2.5):
        return '$$'
    elif (dat['price_new'] < 3.5):
        return '$$$'
    elif (dat['price_new'] >= 3.5):
        return '$$$$'
    else:
        return np.nan
```

```python
dat3['price_final'] = dat3.apply(flg2, axis=1)
dat3
```

```python
final['price'] = dat3['price_final']
```
