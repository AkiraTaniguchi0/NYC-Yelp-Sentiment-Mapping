```python
import timeit
start = timeit.default_timer()

labels = []
scores = []

for i in cleaned['clean_text']:
  seq = i[:512]
  temp = classifier5(seq)
  labels.append(temp[0]['label'])
  scores.append(temp[0]['score'])


cleaned['labels'] = labels
cleaned['scores'] = scores


stop = timeit.default_timer()
print('Time: ', stop - start)
```

```python
cleaned['scores_adj'] = ((cleaned['scores'] - .5)*2)
```

```python
import numpy as np
cleaned['scores_final'] = np.where(cleaned['labels']=='LABEL_1', cleaned['scores_adj'], cleaned['scores_adj']*-1)
```

```python
cleaned['labels_final'] = np.where(cleaned['labels']=='LABEL_1', 'Positive', 'Negative')
```
